```python
from __future__ import print_function
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
M1 = np.loadtxt("txt_files_py/posGtNico25deg.txt") 
M2 = np.loadtxt("txt_files_py/posGtStandard.txt")
print('M1 shape:', M1.shape)
print('M2 shape:', M2.shape)

# Put in seconds
M1[:,0] = M1[:,0]*pow(10, -10)
M2[:,0] = M2[:,0]*pow(10, -10)
# Initialize at 0sec
if M1[0,0]<M2[0,0]:
    M1[:,0] = M1[:,0] - M1[0,0]
    M2[:,0] = M2[:,0] - M1[0,0]
else:
    M1[:,0] = M1[:,0] - M2[0,0]
    M2[:,0] = M2[:,0] - M2[0,0]
```

```python
plt.plot(M1[:,0], M1[:,1], alpha=0.5, label='Rabany-Jiang')
plt.plot(M2[:,0], M2[:,1], alpha=0.5, label='Standard')
plt.legend()
```

```python
plt.plot(M1[:,0], M1[:,2], alpha=0.5, label='Rabany-Jiang')
plt.plot(M2[:,0], M2[:,2], alpha=0.5, label='Standard')
plt.legend()
```
